```python
# Loading Rrequired Packages
import os #Interacting with the Operating system.
import json # For working with JSON files
import csv # For working with CSV files
from pprint import pprint # For printing Python data structures in an organized manner
```

```python
# Function to load data from CSV, JSON, and TXT files
def load_data():
    sales_data = {}
    product_details = {}
    product_descriptions = {}

    # Load sales data from CSV
    csv_file = r'D:\UPGRAD\mainfolder\sales_data.csv'
    with open(csv_file, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            sales_data[row['Product_SKU']] = {day: int(row[day]) for day in reader.fieldnames[1:]}

    # Load product details from JSON files
    details_folder = r'D:\UPGRAD\mainfolder\product_details'
    for filename in os.listdir(details_folder):
        if filename.endswith('.json'):
            with open(os.path.join(details_folder, filename), 'r') as json_file:
                product_details[filename.split('.')[0]] = json.load(json_file)

    # Load product descriptions from TXT files
    descriptions_folder = r'D:\UPGRAD\mainfolder\product_description'
    for filename in os.listdir(descriptions_folder):
        if filename.endswith('.txt'):
            with open(os.path.join(descriptions_folder, filename), 'r') as txt_file:
                product_descriptions[filename.split('.')[0]] = txt_file.read()

    return sales_data, product_details, product_descriptions

# Call the function to load data
sales_data, product_details, product_descriptions = load_data()
```

```python
# Display sales_data, product_details, and product_descriptions dictionaries

def display_data():
    print("Sales Data:")
    pprint(sales_data)
    print("\nProduct Details:")
    pprint(product_details)
    print("\nProduct Descriptions:")
    pprint(product_descriptions)

# Create a list of product SKUs
product_skus = list(product_details.keys())

# Display data for a specific product using its SKU
def display_product_info(product_sku):
    print(f"\nProduct SKU: {product_sku}")
    print("Sales Data:")
    pprint(sales_data.get(product_sku, "Product SKU not found in sales data."))
    print("\nProduct Details:")
    pprint(product_details.get(product_sku, "Product SKU not found in product details."))
    print("\nProduct Description:")
    pprint(product_descriptions.get(product_sku, "Product SKU not found in product descriptions."))

# Display the length of dictionaries
def display_lengths():
    print(f"\nLength of Sales Data Dictionary: {len(sales_data)}")
    print(f"Length of Product Details Dictionary: {len(product_details)}")
    print(f"Length of Product Descriptions Dictionary: {len(product_descriptions)}")

# Call functions to display data
def display():
    display_data()
    display_product_info(product_skus[0])  # Displaying data for the first product in the list
    display_lengths()
```

```python
# Function to add sales data for Acer Aspire 3
def add_sales_data():
    new_product_sku = input("Enter Product SKU You Want To Add: ")
    sales_data[new_product_sku] = []
    for i in range(14):
        sales = int(input(f"Enter sales data for Day {i+1}: "))
        sales_data[new_product_sku].append(sales)

# Function to add product details for Acer Aspire 3
def add_product_details():
    new_product_sku = input("Enter Product SKU You Want To Add: ")
    product_details[new_product_sku] = {
        "Product Name": input("Enter Product Name"),
        "Brand": input("Enter Brand"),
        "Model": input("Enter Model"),
        "Specifications": input("Enter product specifications: "),
        "Price": input("Enter product price: "),
        "Availability": input("Enter product availability: ")
    }

# Function to add product description for Acer Aspire 3
def add_product_description():
    new_product_sku = input("Enter Product SKU You Want To Add: ")
    product_descriptions[new_product_sku] = input("Enter product description: ")

# Function to call all three functions to add data for Acer Aspire 3
def create():
    add_sales_data()
    add_product_details()
    add_product_description()
```

```python
# Function to display sales data for Acer Aspire 3
def display_sales_data():
    product_sku = input("Enter Product SKU You Want To Read: ")
    if product_sku in sales_data:
        print(f"\nSales Data for Product SKU {product_sku}:")
        pprint(sales_data[product_sku])
    else:
        print(f"\nSales Data for Product SKU {product_sku} not found.")

# Function to display product details for Acer Aspire 3
def display_product_details():
    product_sku = input("Enter Product SKU You Want To Read: ")
    if product_sku in product_details:
        print(f"\nProduct Details for Product SKU {product_sku}:")
        pprint(product_details[product_sku])
    else:
        print(f"\nProduct Details for Product SKU {product_sku} not found.")

# Function to display product description for Acer Aspire 3
def display_product_description():
    product_sku = input("Enter Product SKU You Want To Read: ")
    if product_sku in product_descriptions:
        print(f"\nProduct Description for Product SKU {product_sku}:")
        pprint(product_descriptions[product_sku])
    else:
        print(f"\nProduct Description for Product SKU {product_sku} not found.")

# Function to call all three functions to display data for Acer Aspire 3
def read():
    display_sales_data()
    display_product_details()
    display_product_description()
```

```python
# Function to update sales data for Acer Aspire 3
def update_sales_data():
    product_sku = input("Enter Product SKU You Want To Update: ")
    if product_sku in sales_data:
        new_sales_data = []
        for i in range(14):
            sales = int(input(f"Enter updated sales data for Day {i+1}: "))
            new_sales_data.append(sales)
        sales_data[product_sku] = new_sales_data
        print(f"\nUpdated Sales Data for Product SKU {product_sku}:")
        pprint(sales_data[product_sku])
    else:
        print(f"\nProduct SKU {product_sku} not found in sales data.")

# Function to update product details for Acer Aspire 3
def update_product_details():
    product_sku = input("Enter Product SKU You Want To Update: ")
    if product_sku in product_details:
        product_details[product_sku]["Product Name"] = input("Enter updated Product Name: ")
        product_details[product_sku]["Brand"] = input("Enter updated Brand: ")
        product_details[product_sku]["Model"] = input("Enter updated Model: ")
        product_details[product_sku]["Specifications"] = input("Enter updated product specifications: ")
        product_details[product_sku]["Price"] = input("Enter updated product price: ")
        product_details[product_sku]["Availability"] = input("Enter updated product availability: ")
        print(f"\nUpdated Product Details for Product SKU {product_sku}:")
        pprint(product_details[product_sku])
    else:
        print(f"\nProduct SKU {product_sku} not found in product details.")

# Function to update product description for Acer Aspire 3
def update_product_description():
    product_sku = input("Enter Product SKU You Want To Update: ")
    if product_sku in product_descriptions:
        product_descriptions[product_sku] = input("Enter updated product description: ")
        print(f"\nUpdated Product Description for Product SKU {product_sku}:")
        pprint(product_descriptions[product_sku])
    else:
        print(f"\nProduct SKU {product_sku} not found in product descriptions.")

# Function to call all three functions to update data for Acer Aspire 3
def update():
    update_sales_data()
    update_product_details()
    update_product_description()
```

```python
# Function to delete data for Acer Aspire 3
def delete():
    product_sku = input("Enter Product SKU to delete: ")
    
    # Deleting from sales_data dictionary
    if product_sku in sales_data:
        del sales_data[product_sku]
        print(f"\nDeleted Sales Data for Product SKU {product_sku}.")
    else:
        print(f"\nProduct SKU {product_sku} not found in sales data.")

    # Deleting from product_details dictionary
    if product_sku in product_details:
        del product_details[product_sku]
        print(f"\nDeleted Product Details for Product SKU {product_sku}.")
    else:
        print(f"\nProduct SKU {product_sku} not found in product details.")

    # Deleting from product_descriptions dictionary
    if product_sku in product_descriptions:
        del product_descriptions[product_sku]
        print(f"\nDeleted Product Description for Product SKU {product_sku}.")
    else:
        print(f"\nProduct SKU {product_sku} not found in product descriptions.")
```

```python
def CRUD():
    # displaying the existing data
    display()
    
    # Call the create function to add data for Acer Aspire 3
    create()

    # Call the read function to display data for Acer Aspire 3
    read()

    # Call the update function to update data for Acer Aspire 3
    update()

    # Call the delete function to delete data for Acer Aspire 3
    delete()
```
